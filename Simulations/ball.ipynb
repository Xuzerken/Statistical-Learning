```python
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def UniSphere(n, d):
    pts = np.random.randn(n, d)
    for i in range(n):
        pts[i] = pts[i]/np.linalg.norm(pts[i])
    return pts
```

```python
pts = UniSphere(200, 50)
lines = UniSphere(5, 50)
```

```python
for i in range(1,6):
    pj = np.dot(pts, lines[i-1])
    plt.figure(i)
    sns.violinplot(pj)
    plt.grid()
    plt.savefig("images/ball/projection"+str(i)+".png")
```
