```python
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
pts = []
for i in range(30):
    pts.append(2*(np.random.random(100)-0.5))
pts = np.array(pts)
pts
```

```python
from scipy.spatial.distance import pdist
from scipy.spatial.distance import squareform
distA = pdist(pts,metric='euclidean')
distB = squareform(distA)
```

```python
print(distA.var())
print(distA.mean())
```

```python
plt.hist(distA)
plt.grid()
plt.savefig("images/distance/hist.png")
```

```python
sns.violinplot(distA)
plt.grid()
plt.savefig("images/distance/violin.png")
```

```python
plt.figure(figsize=(10, 10))
hm = sns.heatmap(distB,cbar=True,cmap="GnBu",annot=False,square=True,fmt='.2f',annot_kws={'size': 15})
plt.savefig("images/distance/hm.png")
```

```python
cosA = 1- pdist(pts,metric='cosine')
cosB = squareform(cosA)
```

```python
plt.hist(cosA)
plt.grid()
plt.savefig("images/cos/hist.png")
```

```python
sns.violinplot(cosA)
plt.grid()
plt.savefig("images/cos/violin.png")
```

```python
plt.figure(figsize=(10, 10))
hm = sns.heatmap(cosB,cbar=True,cmap="GnBu",annot=False,square=True,fmt='.2f',annot_kws={'size': 15})
plt.savefig("images/cos/hm.png")
```
